```python
cursor.execute('use faculdade')
```

```python
cursor.execute('create table usuarios(id_usuario int auto_increment primary key, nome_usuario varchar(100), cpf varchar (50), telefone varchar(100))')
```

```python
sql = "INSERT INTO usuarios (nome_usuario, cpf, telefone) VALUES (%s, %s, %s)"
val = [
    ("John", "52292345678", "11998765432"),
  ("Deivid", "52348902134", "12965431212")
]

cursor.executemany(sql, val)

conexao.commit()

print(cursor.rowcount, "record inserted.")
```

```python
cursor.execute("SELECT * FROM usuarios")

myresult = cursor.fetchall()

for x in myresult:
  print(x)
```

```python
cursor.execute('create table cursos(id_curso int auto_increment primary key, nomecurso varchar(100), tipo varchar (50), preco varchar(100))')
```

```python
sql = "INSERT INTO cursos (nomecurso, tipo, preco) VALUES (%s, %s, %s)"
val = [
    ("Mecânica", "Technology", "R$4.800,00"),
  ("T.I", "Technology", "R$5.000,00")
]

cursor.executemany(sql, val)

conexao.commit()

print(cursor.rowcount, "record inserted.")
```

```python
sql = "SELECT \
  usuarios.nome_usuario AS user, \
  cursos.nomecurso AS favorite \
  FROM usuarios \
  INNER JOIN cursos ON usuarios.id_usuario = cursos.id_curso"

cursor.execute(sql)

myresult = cursor.fetchall()

for x in myresult:
  print(x)
```
